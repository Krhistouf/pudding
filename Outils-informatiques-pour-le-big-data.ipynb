# Fonction `wordcount_map`

- Coder une fonction nommée `wordcount_map` permettant de lire un fichier et de compter 
les mots transformés en minuscules et ne contenant que des lettres de l'alphabet.
- Utiliser le codage `latin-1` pour ouvrir vos fichiers.
- Cette fonction renvoie un dictionnaire dont chaque mot du texte est une clé et son
nombre d'occurences la valeur correspondante.

```python
import re
from collections import Counter

def wordcount_map(textfile):
    with open(textfile, 'r', encoding='UTF-8') as file:
        data = file.read()
        listeMots = re.findall('\w+', data.lower())
        dictionnaire = dict(Counter(listeMots).most_common())
    return dictionnaire
```

```python
wordcount_map(filenames[0])
```

# Fonction `wordcount_reduce`

- Coder une fonction `wordcount_reduce` avec comme argument une liste de dictionnaires calculés chacun par la fonction `wordcount_map`. Cette fonction retourne un dictionnaire 
contenant les mots apparaissant dans **tous** les livres avec leur nombre d'occurences total. 



```python
listDict = [wordcount_map(filenames[i]) for i in range(4)]
listDict
```

```python
from collections import Counter

def wordcount_reduce(listOfDictionnaries):
    dico = Counter({})
    for dic in listOfDictionnaries:
        dico += dic
    return dict(dico)
```

```python
test=wordcount_reduce(listDict)
test
```

- Afficher le nombre de mots différents pour chaque livre
- Afficher les 5 premiers mots les plus courants.

```python
def difWord(textFile):
    with open(textFile, 'r', encoding='UTF-8') as file:
        data = file.read()
    listeMots = re.findall('\w+', data.lower())
    setWords = set(listeMots)
    return len(setWords)
```

```python
for i in range(4):
    print("{}: {}".format(filenames[i], difWord(filenames[i])))
```

```python
import re
from collections import Counter

def wordcount_map_top(textfile, top=5):
    with open(textfile, 'r', encoding='UTF-8') as file:
        data = file.read()
        listeMots = re.findall('\w+', data.lower())
        dictionnaire = dict(Counter(listeMots).most_common(top))
    return dictionnaire
```

```python
for i in range(4):
    print("{}: {}".format(filenames[i], wordcount_map_top(filenames[i])))
```

- Remplacer la boucle en utilisant `map`

```python
print(*map(difWord, filenames))
```

```python
print(*map(wordcount_map_top, filenames))
```

 - Donner la liste des mots communs au 4 livres commencant par `q`. (avec le code)

```python
import re
from collections import Counter

def wordcount_map_q(textfile):
    with open(textfile, 'r', encoding='UTF-8') as file:
        data = file.read()
        listeMots = re.findall('\w+', data.lower())
        dictionnaire = list(Counter(listeMots).most_common())
        listeQ = [elem[0] for elem in dictionnaire if elem[0][0]=='q']
    return listeQ
```
